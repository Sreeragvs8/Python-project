```python
import tkinter as tk

def on_click(key):
    if key == "=":
        try:
            result = eval(entry.get())
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif key == "C":
        entry.delete(0, tk.END)
    else:
        entry.insert(tk.END, key)

# Create the main window
root = tk.Tk()
root.title("Realistic Calculator")

# Center the window on the screen
window_width = 400
window_height = 400
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
x = (screen_width // 2) - (window_width // 2)
y = (screen_height // 2) - (window_height // 2)
root.geometry(f'{window_width}x{window_height}+{x}+{y}')

# Styling constants
BG_COLOR = '#f0f0f0'  # Light gray background color
ENTRY_BG_COLOR = '#ffffff'  # White background color for entry widget
BUTTON_BG_COLOR = '#e0e0e0'  # Light gray background color for buttons
FONT = ('Helvetica', 18)  # Font for buttons and entry widget

# Configure padding for the entire window
root.configure(bg=BG_COLOR, padx=10, pady=10)

# Create a frame to contain all widgets and add a border
frame = tk.Frame(root, borderwidth=10, relief="ridge", bg=BG_COLOR)
frame.pack(padx=10, pady=10)

# Create the Entry widget (where the digits and results will be displayed)
entry = tk.Entry(frame, width=35, borderwidth=2, font=FONT, bg=ENTRY_BG_COLOR, justify="right")
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10, sticky="nsew")

# Define buttons
buttons = [
    ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
    ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
    ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
    ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),
    ('C', 5, 1)  # Center 'C' button in the middle of the last row
]

# Create and place buttons in the grid
for button_text, row, column in buttons:
    if button_text == 'C':
        btn = tk.Button(frame, text=button_text, padx=20, pady=10, font=FONT, bg=BUTTON_BG_COLOR, borderwidth=2, command=lambda key=button_text: on_click(key))
        btn.grid(row=row, column=column, columnspan=2, padx=5, pady=5, sticky="nsew")  # Set columnspan to 2 for 'C' button
    else:
        btn = tk.Button(frame, text=button_text, padx=20, pady=10, font=FONT, bg=BUTTON_BG_COLOR, borderwidth=2, command=lambda key=button_text: on_click(key))
        btn.grid(row=row, column=column, padx=5, pady=5, sticky="nsew")

# Configure row and column weights
for i in range(6):  # Assuming 6 rows
    frame.grid_rowconfigure(i, weight=1)
for i in range(4):  # Assuming 4 columns
    frame.grid_columnconfigure(i, weight=1)

root.mainloop()
```
